## Dealing with Null values 

```python
df.isnull().sum().sum()
```

```python
#Percentage of null values are there
missing_percentages = df.isnull().sum().sort_values(ascending=False)/len(df)
#graph view with missing columns
missing_percentages[missing_percentages!=0].plot(kind="barh")
```

## DROP COLUMNS (80% above missing data )

```python
#REMOVE COLUMNS WITH 80% NULL VALUE
#Removing columns having more than 80% of null values
missing_features = missing_percentages[missing_percentages>0.80].index
df.drop(missing_features, axis=1, inplace=True)
```

```python
#Percentage of null values are there
missing_percentages = df.isnull().sum().sort_values(ascending=False)/len(df)
#graph view with missing columns
missing_percentages[missing_percentages!=0].plot(kind="barh")
```
